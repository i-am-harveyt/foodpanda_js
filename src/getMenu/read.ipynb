```python
import json
import pandas
import base64
```

```python
filePath = "../../../uber_data/uber_menu/2023-12-6/shopLst_21.9414135330001_120.720446352_2023-12-6.csv_2023-12-6.csv"
data = pandas.read_csv(filePath)
data.head()
```

```python
data["menu_decoded"] = data["menu"].apply(
    lambda x: json.loads(base64.b64decode(x).decode("utf-8")))
data.head()
```
